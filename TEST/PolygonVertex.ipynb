# PolygonVertex

```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

from Utils import slit, hexagon, heptagon, octagon, square
```

## slit

```python
import matplotlib.pyplot as plt

# Assuming slit function returns some data to plot
data = slit(4, 4, 0)

print(data.shape)
print(data)

# Plot the data
plt.scatter(data[:, 0], data[:, 1])
plt.plot(data[:, 0], data[:, 1])

for idx in [0, len(data) // 2, len(data) - 1]:
    plt.annotate(str(idx), (data[idx, 0], data[idx, 1]), textcoords="offset points", xytext=(12, -2), ha='right')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.gca().set_axisbelow(True)
plt.show()
```
